```python
import pandas as pd
import numpy as np
Series_A=pd.Series([10,20,30,40,50])
Series_B=pd.Series([40,50,60,70,80])
union=pd.Series(np.union1d(Series_A.values,Series_B.values))
intersect=pd.Series(np.intersect1d(Series_A.values,Series_B.values))
notcommonseries=union[-union.isin(intersect)]
print(notcommonseries)
```

```python
import pandas as pd
import numpy as  np
data={
    'First':['Aryan','Rohan','Riya','Yash','Siddini'],
    'Lastname':['singh','Agarwal','Shah','Batia','Khan'],
    'Type':['Full-time employee','intern','Full-time employee','Part time employee','Full-time employee'],
    'Department':['Administration','Technical','Administration','Technical','Technical'],
    'Yoe':[2,3,5,7,6],
    'Salary':[20,000,5000,10000,20000,]
}
df=pd.DataFrame(data)
print(df)
table=pd.pivot_table(df,index=['Type','Department'],
                     values='Salary',aggfunc=['mean'])
print(table)
```
