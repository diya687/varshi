```python
def Start_with_A(s):
  return s[0]=="A"
fruit=["Apple","banana","pear","Apricot","orange"]
filter_object=filter(Start_with_A,fruit)
print(list(filter_object))
```

```python
def oddeven(x):
  if x%2==0:
    return True
  else:
    return False
list1=[4,5,6,7,8,9]
evenlist=list(filter(oddeven,list1))
print(evenlist)
```
